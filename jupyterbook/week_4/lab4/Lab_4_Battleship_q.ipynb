None of the next functions are called in the cells in which they are defined by default. You may add calls to run the functions inside their cells if you like, or use the ones in **5: Play the game** below. 

`initialize_board()` is given to you in its entirety (minus the doc string above). The operations it performs use a for loop to execute a program designed to generate data for a game board. 

The for loop in `initialize_board()` uses an iterator `_` generated by `range(size)`, an iterable that produces size values. The `_` acts as a placeholder. It will automatically be replaced with `~` as the following two lines are executed. 

Each time the loop runs, a row of `~` symbols is created and appended to the board object. The function returns the complete `size` Ã— `size` grid, called `board`. 

````python
def initialize_board(size):
    """
    # BEGIN SOLUTION
    Initialize a game board for tracking ships or shots.

    :param size: The size of the game board (size x size).
    :return: A grid filled with '~' to represent water.
    # END SOLUTION     
    """
    board = []
    for _ in range(size):
        row = ["~"] * size
        board.append(row)
    return board
````

````python
def place_ship(board, coordinates):
    """
    ...
    """
    # BEGIN SOLUTION
    for row, col in coordinates:
    # END SOLUTION
        # BEGIN SOLUTION
        row -= 1  # Convert to 1-based index
        # END SOLUTION
        # BEGIN SOLUTION
        col -= 1  # Convert to 0-based index
        # END SOLUTION
        board[row][col] = "S"
````

To make it easier to visually parse, `Board`, the object we use to pass data between functions, will be displayed using a unicode character (`â– ` in different colors for ships, hits, and misses). However, inside each of the functions `board` will be defined using the single string characters, `"~"`, `"S"`, `"X"`, and `"M"`. This is becuase python cannot perform operations using unicode characters as objects. However, it can display unicode characters â€” even with formatting options like color â€” to the console, which is how we will display the board as the last step once operations related to ship placement or targeting are performed. 

`display_board()` uses two for loops, 4 logical tests, and a print() call to display a modified version of `board` directly in the console. It generates a blank list which will be used to store data representing the modified version that actually gets printed to the console.

The outer loop iterates through each `row` of `board`, making a new blank list, which we will call `color_dat`. It has exactly  enough entries to store data describing the desired form and color of each individual `element` in that `row`. The inner loop performs a logical test on each `element` in the `row` of `board`, and appends a string of formatting characters to `color_dat` depending on what was found at that `element`. This process continues running until every entry in the matrix has been checked and `color_dat` is a matrix of strings containing exactly the same number of elements as `board`, but in a form that looks much better when output the console. We don't need to worry about retaining this data to perform operations on, so we just print directly to the console (`display_board` has no return line).

A major benefit of using a new blank list populated with formatting data for the printed output is that `board` still exists as an object retaining all of its original data (along with any updates from `place_ship()` or `fire_shot()`). Since `board` is easy to update and maintain, that let's us continually add new data to it without having to reinitialize every time.

```python
    """
    Fire a shot at the board and return the result.

    :param board: The game board.
    :param target: A tuple (row, col) where the shot is fired (1-based indexing).
    :return: The result of the shot: "Hit", "Miss", or "Already Fired".
    """

```

We want to manipulate `board` using indices based on the input tuple `target`. Let's break `target` into individual integers representing the row and column indices. 
- Add a line to define two variables, `row` and `col` which are to store values contained in `target`, replacing the first `...` after the doc string
- We want to allow for human input data, so we will need to convert `row` and `col` to a 1-based indexing system. Replace the `...` on the next two lines
    
Next, we need to determine what is currently occupying the index corresponding to our targeting information and what will happen to that index as a result of a shot being fired there. 

If the target is an `X`, that means we've already shot at that space, so we want to return a message stating `"Already Fired"`. 

If the target contains an `S` that means it contains part of a ship, so we want to update the space with an `X` and return `"Hit"`. 

If the target contains `~` that means it contains water, so we want to update the space with an `M` and return `"Miss"`. 

Since we built the board, we know those are the only possible options, so everything else should be an invalid input, returning `"Invalid Target"`.

- Fill in the 4 logical expressions below to complete `fire_shot()`, replacing 9 sets of `...`

````python
def fire_shot(board, target):
    """
    ...
    """
    # BEGIN SOLUTION
    row, col = target
    row -= 1  # Convert to 1-based index
    col -= 1  # Convert to 1-based index
    # END SOLUTION

    if board[row][col] == "X":
        # BEGIN SOLUTION
        return "Already Fired"
        # END SOLUTION
    elif board[row][col] == "S":
        # BEGIN SOLUTION
        board[row][col] = "X"
        return "Hit"
        # END SOLUTION
    elif board[row][col] == "~":
        # BEGIN SOLUTION
        board[row][col] = "M"
        return "Miss"
        # END SOLUTION
    else:
        return "Invalid Target"   
````

You want to make sure you're not aiming at the same place twice.

Run the following cell to edit the tracking board.

````python
def record_opponent_feedback(tracking_board):
    """
    Prompt the user for a target and result, then update the tracking board.

    :param tracking_board: The board tracking your shots on the opponent's board.
    """
    try:
        # Get input for target coordinates as a tuple
        target_input = input("row,col: ")
        target = tuple(map(int, target_input.split(",")))

        # Get input for result ("Hit" or "Miss")
        result = input("hit or miss?: ").capitalize()
        
        # Update the tracking board
        row, col = target
        row -= 1  # Convert to 0-based index
        col -= 1  # Convert to 0-based index

        if result == "Hit":
            tracking_board[row][col] = "X"  # Mark hit
        elif result == "Miss":
            tracking_board[row][col] = "M"  # Mark miss
        else:
            print("Invalid result. Please enter 'hit' or 'miss'.")
    except ValueError:
        print("Invalid input. Please ensure coordinates are in the format row,col.")

tracking_board = initialize_board(size=10)
````
